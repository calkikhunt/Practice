```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="day", style="smoker", size="size")
```

```python
df1 = pd.DataFrame(np.random.randn(300, 2), columns=["a", "b"])
df1.head()
```

```python
sns.relplot(x="a", y="b", kind="line", data=df1, sort=True)
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)
```
