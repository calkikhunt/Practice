```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
%matplotlib inline
```

```python
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.randn(300, 3), columns=['a','b','c'])
df1.head()
```

```python
df1.iplot(mode="markers")
```

```python
df1.iplot(x="a", y="b", mode="markers")
```

```python
df2 = pd.DataFrame(np.random.rand(300, 3), columns=['a','b','c'])
df2.head()
```

```python
df2.iplot(x="a", y="b", mode="markers")
```

```python
df2.iplot(mode="markers")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.iplot(x="sex", y="survived", kind="bar")
```

```python
cf.set_config_file(theme="ggplot")
titanic.iplot(x="age", y="fare", mode="markers", xTitle="age", yTitle="fare")
```
