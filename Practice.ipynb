```python
import numpy as np
import statsmodels.api as sm
import pylab as py

a = np.random.normal(0, 1, 100)
sm.qqplot(a, line='45')
py.show()
```

```python
import seaborn as sns
b = np.random.lognormal(0, 1, 100)
sns.displot(b, discrete=True, kde=True)
```

```python
sm.qqplot(np.log(b), line='45')
py.plot()
```
